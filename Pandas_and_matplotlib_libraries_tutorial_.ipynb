Tutorial in pandas and matplotlib libraries

The problem is to graph a horizontal bar for the top 15 programming languages in given csv file.

Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read csv file

```python
data =pd.read_csv("/content/data-2.csv")
data. head()
```

See the information of data frame

```python
data.info()
```

Split every language in each row

```python
language = data['LanguagesWorkedWith'].str.split(";")
```

Creat an empty list

```python
language_list =[]
```

Extend language in the last list

```python
for lang in language:

       language_list.extend(lang)
```

Creat pandas series for count of language

```python
language_counts = pd.Series(language_list).value_counts()
```

Identify the top 15 languages

```python
top_languages = language_counts.head(15)
```

Graph the top 15 languages in horizontal bar

```python

plt.figure(figsize=(10, 7))
top_languages.plot(kind='barh')
plt.title('Top 15 Languages Used')
plt.show()
```
